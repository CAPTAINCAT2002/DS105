Câu 9. Sử dụng Nearest neighbors để điền các giá trị bị thuyết.

```python
df_1=df
```

```python
df_1
```

```python
#lọc ra những features có kiểu dữ liệu số
filt = df_1.apply(lambda x: x.dtype==np.int64 or x.dtype==np.float64)
df_2 = df_1[df_1.columns[filt]]
df_2
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=4, metric='nan_euclidean')
df_trans = imputer.fit_transform(df_2)
fill_data = pd.DataFrame(df_trans, columns=df_2.columns)
df_1[df_1.columns[filt]] = fill_data
```

```python
df_1.isna().sum()
```

```python
df_1.to_csv('df_cau09.csv')
```

Câu 12. Chuẩn hóa "length", "width" và "height" trong phạm vi [0,1].

```python
df_c12=df
```

```python
df_c12[['length','width','height']]
```

```python
df_c12['length'] = (df_c12['length'] - df_c12['length'].min())/(df_c12['length'].max() - df_c12['length'].min())
df_c12['width']= (df_c12['width'] - df_c12['width'].min())/(df_c12['width'].max() - df_c12['width'].min())
df_c12['height']= (df_c12['height'] - df_c12['height'].min())/(df_c12['height'].max() - df_c12['height'].min())
```
